### Array Math

```python
 x = np.array(([1,2], [3,4]), dtype=np.float64)
 y = np.array(([5,6], [7,8]), dtype=np.float64)
```

```python
print(x + y)
print(np.add(x, y))
```

```python
print(x - y)
print(np.subtract(x, y))
```

```python
print(x * y)
print(np.multiply(x,y))
```

```python
print(np.sqrt(x))
```

```python
v = np.array([9, 10])
w = np.array([11, 12])
print(np.dot(v,w))
print(x @ v)
```
